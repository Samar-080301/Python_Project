```python
import pandas as pd
df = pd.read_csv(r'C:\Users\samar\OneDrive\Desktop\MAINY\projects\Pandas-Data-Science-Tasks\SalesAnalysis\all_data_copy.csv')
```

```python
df.head()
```

```python
df2 = df[df['Product']!='Product']
df = df2.dropna(how='all')
df['Date']= df['Order Date'].str[3:5]
df['Month']= df['Order Date'].str[:2]
df['Year']= df['Order Date'].str[6:8]
df.rename(columns={"Price Each":"Rate"},inplace=True)
df.sort_values(by=['Order ID'],ascending=True)
df['Rate']=pd.to_numeric(df['Rate'])
df['Quantity Ordered']=pd.to_numeric(df['Quantity Ordered'])
df['Price']=df['Rate']*df['Quantity Ordered']
df.head()
```

```python
m=df.groupby('Month').sum()['Quantity Ordered']
```
